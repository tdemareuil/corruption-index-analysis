### Clean environment variables

#### CO2

```python
co2 = pd.read_csv('Environment/CO2 Emission Estimates.csv', header=1)[['Unnamed: 1', 'Value', 'Year', 'Series']]
co2 = co2[co2['Year'].astype(int)>2011]
```

```python
co2_total = co2[co2.Series=='Emissions (thousand metric tons of carbon dioxide)'].drop(columns=['Series'])
co2_total = co2_total.rename(columns={'Value':'CO2 emissions (thousand tons)', 'Unnamed: 1':'Country'}).reset_index(drop=True)
co2_total = co2_total.pivot(index='Country', columns=['Year']).reset_index()
co2_total.columns = ['Country', 'CO2_total_2015', 'CO2_total_2016', 'CO2_total_2017']
co2_total
```

```python
co2_capita = co2[co2.Series=='Emissions per capita (metric tons of carbon dioxide)'].drop(columns=['Series'])
co2_capita = co2_capita.rename(columns={'Value':'Emissions per capita (CO2 tons)', 'Unnamed: 1':'Country'}).reset_index(drop=True)
co2_capita = co2_capita.pivot(index='Country', columns=['Year']).reset_index()
co2_capita.columns = ['Country', 'CO2_capita_2015', 'CO2_capita_2016', 'CO2_capita_2017']
co2_capita
```

```python
co2 = co2_capita.merge(co2_total, on='Country', how='inner')
co2
```

```python
co2.isna().sum()
```

```python
co2.to_csv('co2_emissions.csv')
```

#### Poverty

```python
poverty = pd.read_csv('Social/raw/Poverty_Stats_csv/PovStatsData.csv')[['Country Name', 'Indicator Name', '2012', '2013', '2014', '2015', '2016', '2017']]
poverty
```

```python
poverty['Indicator Name'].unique()
```

```python
keep = ['Income share held by highest 10%',
'Annualized growth in per capita real survey mean consumption or income, total population (%)',
'Annualized growth in per capita real survey mean consumption or income, top 10% (%)',
'Annualized growth in per capita real survey mean consumption or income, bottom 40% (%)',
'Gini index (World Bank estimate)',
'Median daily per capita income or consumption expenditure (2011 PPP)',
        'Multidimensional poverty, Drinking water (% of population deprived)',
        'Multidimensional poverty, Electricity (% of population deprived)',
        'Poverty headcount ratio at $1.90 a day (2011 PPP) (% of population)',
        'Poverty headcount ratio at $5.50 a day (2011 PPP) (% of population)',
'Rural poverty headcount ratio at national poverty lines (% of rural population), including noncomparable values',
       'Survey mean consumption or income per capita, bottom 40% (2011 PPP $ per day)',
       'Survey mean consumption or income per capita, top 10% (2011 PPP $ per day)',
        'Survey mean consumption or income per capita, total population (2011 PPP $ per day)',
       'Urban poverty headcount ratio at national poverty lines (% of urban population), including noncomparable values']
```

```python
poverty = poverty[poverty['Indicator Name'].isin(keep)]
poverty
```

```python
renames = {'Income share held by highest 10%':'income_share_top10%',
'Annualized growth in per capita real survey mean consumption or income, total population (%)':'income_growth_total',
'Annualized growth in per capita real survey mean consumption or income, top 10% (%)':'income_growth_top10%',
'Annualized growth in per capita real survey mean consumption or income, bottom 40% (%)':'income_growth_bottom40%',
'Gini index (World Bank estimate)':'gini_index',
'Median daily per capita income or consumption expenditure (2011 PPP)':'median_daily_income_ppp',
        'Multidimensional poverty, Drinking water (% of population deprived)':'no_water_%',
        'Multidimensional poverty, Electricity (% of population deprived)':'no_electricity_%',
        'Poverty headcount ratio at $1.90 a day (2011 PPP) (% of population)':'extreme_poverty_%_$1.90',
        'Poverty headcount ratio at $5.50 a day (2011 PPP) (% of population)':'moderate_poverty_%_$5.50',
'Rural poverty headcount ratio at national poverty lines (% of rural population), including noncomparable values':'poverty_%_rural',
       'Survey mean consumption or income per capita, bottom 40% (2011 PPP $ per day)':'mean_daily_income_ppp_bottom40%',
       'Survey mean consumption or income per capita, top 10% (2011 PPP $ per day)':'mean_daily_income_ppp_top10%',
        'Survey mean consumption or income per capita, total population (2011 PPP $ per day)':'mean_daily_income_ppp',
       'Urban poverty headcount ratio at national poverty lines (% of urban population), including noncomparable values':'poverty_%_urban'}
```

```python
poverty['Indicator Name'] = poverty['Indicator Name'].apply(lambda x: renames[x])
poverty
```

```python
poverty = poverty.pivot(index='Country Name', columns=['Indicator Name']).reset_index()
poverty = poverty.swaplevel(i=-1, j=-2, axis=1)
poverty.columns = ['_'.join(col).strip() for col in poverty.columns.values]
#poverty.columns=poverty2.columns.to_flat_index()
poverty = poverty.reindex(sorted(poverty.columns), axis=1)
poverty
```

```python
x = poverty.count().to_frame().reset_index()
x
```

```python
colkeep = list(x[x[0]>50]['index'])
colkeep
```

```python
poverty = poverty[colkeep]
poverty = poverty.rename(columns={"_Country Name": "Country"})
poverty
```

```python
poverty[['gini_index_2012',
         'gini_index_2013','gini_index_2014','gini_index_2015',
         'gini_index_2016','gini_index_2017']] = poverty[['gini_index_2012',
         'gini_index_2013','gini_index_2014','gini_index_2015',
         'gini_index_2016','gini_index_2017']].fillna(method ='ffill', axis=1).fillna(method='bfill', axis=1)
poverty
```

```python
gini = pd.read_clipboard() # from https://data.worldbank.org/indicator/SI.POV.GINI?end=1997&start=1997&view=bar
```

```python
gini
```

```python
import numpy as np
for idx, row in poverty.iterrows():
    if np.isnan(row['gini_index_2012']):
        country = row['Country']
        value = list(gini[gini['Country'] == country]['Most Recent Value'])
        if len(value)>0:
            value = value[0]
        else:
            value=np.nan
        poverty['gini_index_2012'][idx] = value
```

```python
poverty[['gini_index_2012',
         'gini_index_2013','gini_index_2014','gini_index_2015',
         'gini_index_2016','gini_index_2017']] = poverty[['gini_index_2012',
         'gini_index_2013','gini_index_2014','gini_index_2015',
         'gini_index_2016','gini_index_2017']].fillna(method ='ffill', axis=1)
poverty
```

```python
poverty.to_csv('Social/clean/poverty.csv')
```
